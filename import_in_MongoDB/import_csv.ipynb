```python
import csv
from pymongo import MongoClient
import json
from datetime import datetime
from bson import ObjectId, Binary

# Connessione a MongoDB
client = MongoClient('mongodb://localhost:27017/')
db = client['ldbc']
```

```python
#Comment
collection = db['Comment']

def converti_valori_generico(riga):
    for chiave, valore in riga.items():
        val = valore.strip()

        # Valori vuoti o null
        if val.lower() in ('null', ''):
            riga[chiave] = None
            continue

        # Booleani
        if val.lower() == 'true':
            riga[chiave] = True
            continue
        elif val.lower() == 'false':
            riga[chiave] = False
            continue

        # Interi
        try:
            if val.isdigit() or (val[0] == '-' and val[1:].isdigit()):
                riga[chiave] = int(val)
                continue
        except:
            pass

        # Float
        try:
            if "." in val or "," in val:
                val_num = val.replace(",", ".")
                riga[chiave] = float(val_num)
                continue
        except:
            pass

        # Date (vari formati)
        for fmt in ("%Y-%m-%d", "%Y/%m/%d", "%d/%m/%Y", "%Y-%m-%dT%H:%M:%S"):
            try:
                riga[chiave] = datetime.strptime(val, fmt)
                break
            except:
                pass
        else:
            # Tenta come array JSON
            try:
                obj = json.loads(val)
                if isinstance(obj, list):
                    riga[chiave] = obj
                    continue
                elif isinstance(obj, dict):
                    riga[chiave] = obj
                    continue
            except:
                pass

            # ObjectId?
            try:
                if len(val) == 24:
                    riga[chiave] = ObjectId(val)
                    continue
            except:
                pass

            # Lascia come stringa
            riga[chiave] = val

    return riga

with open('dynamic_file_csv/comment.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file, delimiter='|')
    dati = [converti_valori_generico(riga) for riga in reader]

collection.insert_many(dati)
print("Importazione completata!")
```

```python
# Connessione a MongoDB
client = MongoClient('mongodb://localhost:27017/')
db = client['ldbc']

#place_isPartOf_place
place_collection = db['Place']
collection = db['place_isPartOf_place']

# Creiamo una mappatura dagli ID numerici agli ObjectId
id_mapping = {}
places = place_collection.find({}, {"_id": 1, "id": 1})

for place in places:
    # Il campo "id" contiene l'ID numerico che corrisponde agli ID nel CSV
    if "id" in place:
        # Convertiamo l'ID numerico in stringa per sicurezza, poichÃ© nel CSV saranno stringhe
        id_mapping[str(place["id"])] = place["_id"]

print(f"Mappatura creata con {len(id_mapping)} posti")

# Contatori per il report
inserts = 0
errors = 0
not_found = 0

# Leggi il file CSV
with open('static_file_csv/place_isPartOf_place.csv', 'r') as file:
    csv_reader = csv.reader(file, delimiter='|')
    
    for row in csv_reader:
        if len(row) == 2:
            place1_id = row[0].strip()
            place2_id = row[1].strip()
            
            # Verifica se entrambi gli ID sono nella mappatura
            if place1_id in id_mapping and place2_id in id_mapping:
                place1_obj_id = id_mapping[place1_id]
                place2_obj_id = id_mapping[place2_id]
                
                try:
                    # Inserisci la relazione nella collezione delle relazioni
                    relation = {
                        "place1_id": place1_obj_id,
                        "place2_id": place2_obj_id,
                    }
                    
                    result = collection.insert_one(relation)
                    inserts += 1
                    
                except Exception as e:
                    print(f"Errore nell'inserimento per gli ID {place1_id} e {place2_id}: {str(e)}")
                    errors += 1
            else:
                if place1_id not in id_mapping:
                    print(f"ID non trovato nella mappatura: {place1_id}")
                if place2_id not in id_mapping:
                    print(f"ID non trovato nella mappatura: {place2_id}")
                not_found += 1

print(f"Inserimenti completati: {inserts}")
print(f"ID non trovati: {not_found}")
print(f"Errori riscontrati: {errors}")
client.close()


```

```python
# Connessione a MongoDB
client = MongoClient('mongodb://localhost:27017/')
db = client['ldbc']

#comment_hasTag_tag
comment_hasTag_tag_collection = db['comment_hasTag_tag']
comment_collection = db['Comment']
tag_collection = db['Tag']

# Mappature ID -> ObjectId
comment_mapping = {}
tag_mapping = {}

# Crea mappatura per Comment
comments = comment_collection.find({}, {"_id": 1, "id": 1})
for comment in comments:
    if "id" in comment:
        comment_mapping[str(comment["id"])] = comment["_id"]

# Crea mappatura per Tag
tags = tag_collection.find({}, {"_id": 1, "id": 1})
for tag in tags:
    if "id" in tag:
        tag_mapping[str(tag["id"])] = tag["_id"]

print(f"Mappatura Comment creata con {len(comment_mapping)} elementi")
print(f"Mappatura Tag creata con {len(tag_mapping)} elementi")

# Contatori per il report
inserts = 0
errors = 0
not_found = 0

# Leggi il file CSV
with open('static_file_csv/comment_hasTag_tag.csv', 'r', encoding='utf-8') as file:
    csv_reader = csv.reader(file, delimiter='|')
    
    for row in csv_reader:
        raw_date = row[0].strip()
        comment_id = row[1].strip()
        tag_id = row[2].strip()

        # Parsing data robusto
        creation_date = None
        for fmt in ("%Y-%m-%d", "%Y/%m/%d", "%d/%m/%Y", "%Y-%m-%dT%H:%M:%S", "%m/%d/%Y %H:%M:%S"):
            try:
                creation_date = datetime.strptime(raw_date, fmt)
                break
            except:
                pass
        
        if not creation_date:
            print(f"Formato data non riconosciuto: {raw_date}")
            errors += 1
            continue
        
        if comment_id in comment_mapping and tag_id in tag_mapping:
            comment_obj_id = comment_mapping[comment_id]
            tag_obj_id = tag_mapping[tag_id]

            try:
                relation = {
                    "creationDate": creation_date,
                    "comment_id": comment_obj_id,
                    "tag_id": tag_obj_id
                }

                result = comment_hasTag_tag_collection.insert_one(relation)
                inserts += 1

            except Exception as e:
                print(f"Errore inserimento relazione {comment_id}-{tag_id}: {e}")
                errors += 1
        else:
            if comment_id not in comment_mapping:
                print(f"Comment ID non trovato: {comment_id}")
            if tag_id not in tag_mapping:
                print(f"Tag ID non trovato: {tag_id}")
            not_found += 1

print(f"Inserimenti completati: {inserts}")
print(f"ID non trovati: {not_found}")
print(f"Errori riscontrati: {errors}")

client.close()
```

```python
# Connessione a MongoDB
client = MongoClient('mongodb://localhost:27017/')
db = client['ldbc']

#organisation_isLocatedIn_place
organisation_collection = db['Organisation']
place_collection = db['Place']
collection = db['organisation_isLocatedIn_place']

# Create mappings from numeric IDs to ObjectIds
org_id_mapping = {}
organisations = organisation_collection.find({}, {"_id": 1, "id": 1})

for org in organisations:
    if "id" in org:
        # Convert the numeric ID to string for safety, as in CSV they'll be strings
        org_id_mapping[str(org["id"])] = org["_id"]

place_id_mapping = {}
places = place_collection.find({}, {"_id": 1, "id": 1})

for place in places:
    if "id" in place:
        place_id_mapping[str(place["id"])] = place["_id"]

# Counters for reporting
inserts = 0
errors = 0
not_found = 0

# Read the CSV file
with open('static_file_csv/organisation_isLocatedIn_place.csv', 'r') as file:
    csv_reader = csv.reader(file, delimiter='|')
    
    for row in csv_reader:
        if len(row) == 2:
            org_id = row[0].strip()
            place_id = row[1].strip()
            
            # Verify if both IDs are in the mappings
            if org_id in org_id_mapping and place_id in place_id_mapping:
                org_obj_id = org_id_mapping[org_id]
                place_obj_id = place_id_mapping[place_id]
                
                try:
                    # Insert the relationship into the relations collection
                    relation = {
                        "organisation_id": org_obj_id,
                        "place_id": place_obj_id,
                    }
                    
                    result = collection.insert_one(relation)
                    inserts += 1
                    
                except Exception as e:
                    print(f"Error inserting for Organisation ID {org_id} and Place ID {place_id}: {str(e)}")
                    errors += 1
            else:
                if org_id not in org_id_mapping:
                    print(f"Organisation ID not found in mapping: {org_id}")
                if place_id not in place_id_mapping:
                    print(f"Place ID not found in mapping: {place_id}")
                not_found += 1

print(f"Insertions completed: {inserts}")
print(f"IDs not found: {not_found}")
print(f"Errors encountered: {errors}")
client.close()
```
